```python
import time, os, sys, subprocess, shutil
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sp_grep(file, match):
    o = subprocess.run('grep "{}" {}'.format(match, file), check=True, shell=True, stdout=subprocess.PIPE, universal_newlines=True)
    return o.stdout
def ae_loss_split(grepopt, nat):
    epochs = [i.split('||')[0].strip()[-2:].strip() for i in grepopt.split('\n') if i != '']
    tails = [i.split(',')[-1].split('.')[0] for i in grepopt.split('\n') if i != '']
    return (epochs, tails)
```

```python
f = '/home/awills/Documents/Research/swxcd/train/test_subset2/ptscan/2022-08-04_18:48:30.927843.log'
o = [float(i.split(',')[1]) for i in sp_grep(f, 'AE loss').split('\n') if i]
lst = []
tl = []
nat = 7
for idx, num in enumerate(o):
    tl.append(num)
    if (idx > 0) and ((idx+1)%nat == 0):
        tl = sorted(tl)
        lst.append(tl)
        tl = []
```
