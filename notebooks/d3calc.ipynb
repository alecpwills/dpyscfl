# Calculating the D3 Dispersion Corrections to Datasets
Calculating the PBE and HF parametrizations

### G2/97

```python
g2 = read('/home/awills/Documents/Research/swxcd/seawulf/g297/caresults.traj', ':')
g2_pbe = []
g2_hf = []
g2d3_pbe = []
g2d3_hf = []
d3c_pbe = []
d3c_hf = []
for idx, at in enumerate(g2):
    print(idx, at, at.info)
    form = at.get_chemical_formula()
    print('=======================')
    print("{} -- {}".format(idx, form))
    name, mol = ase_atoms_to_mol(at)
    mfp = dft.UKS(mol)
    mfp.xc = 'pbe'
    mfp.grids.level = 5
    mfp.kernel()
    d3p = dftd3.dftd3(mfp)
    print("Calculating D3 dispersion correction -- PBE")
    d3p.kernel()
    g2_pbe.append(mfp.e_tot)
    g2d3_pbe.append(d3p.e_tot)
    d3c_pbe.append(d3p.e_tot - mfp.e_tot)
    mfh = scf.UHF(mol)
    mfh.kernel()
    d3h = dftd3.dftd3(mfh)
    print("Calculating D3 dispersion correction -- HF")
    d3h.kernel()
    g2_hf.append(mfh.e_tot)
    g2d3_hf.append(d3h.e_tot)
    d3c_hf.append(d3h.e_tot - mfh.e_tot)
    print("D3 CALCULATION COMPLETE: {} -- {}".format(idx, form))
    print("D3 Correction -- PBE: {} - {} = {}".format(d3p.e_tot, mfp.e_tot, d3p.e_tot - mfp.e_tot))
    print("D3 Correction -- HF: {} - {} = {}".format(d3h.e_tot, mfh.e_tot, d3h.e_tot - mfh.e_tot))
    print('=======================')
```
